**Null Hypothesis (H0):**
The distribution of match results (wins, draws, losses) is the same for Barcelona's home games at Camp Nou and Montjuic.

**Alternative Hypothesis (H1):**
The distribution of match results (wins, draws, losses) is different for Barcelona's home games at Camp Nou and Montjuic.


# Hypthesis Testing

## 1. Results Distribution

```python
import scipy.stats as stats

# Creating a contingency table
contingency_table = pd.DataFrame({
    'Camp Nou': [3, 1, 15],  # Draws, Losses, Wins
    'Montjuic': [1, 3, 15]   # Draws, Losses, Wins
}, index=['d', 'l', 'w'])

# Performing Chi-Square test
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print(f"Chi-Square Test: chi2 = {chi2}, p-value = {p}")
```

**Chi-Square Test Result:**
- Chi-Square Statistic (chi2): 2.0
- p-value: 0.36787944117144245

**Interpretation:**
The p-value is greater than 0.05, suggesting that there is no statistically significant difference in the distribution of match results between Camp Nou and Montjuic. Therefore, we fail to reject the null hypothesis.


## 2. Proportion of Losses 

`**Null Hypothesis (H0):**
The proportion of losses is the same for Barcelona's home games at Camp Nou and Montjuic.

**Alternative Hypothesis (H1):**
The proportion of losses is different for Barcelona's home games at Camp Nou and Montjuic.
`

```python
# Creating a contingency table for losses
contingency_table = pd.DataFrame({
    'Camp Nou': [1, 18],  # Losses, Non-losses
    'Montjuic': [3, 16]   # Losses, Non-losses
}, index=['Losses', 'Non-losses'])

# Performing Chi-Square test
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print(f"Chi-Square Test: chi2 = {chi2}, p-value = {p}")
```

**Chi-Square Test Result for Losses:**
- Chi-Square Statistic (chi2): 0.28
- p-value: 0.60

**Interpretation:**
The p-value is greater than 0.05, suggesting that there is no statistically significant difference in the proportion of losses between Camp Nou and Montjuic. Therefore, we fail to reject the null hypothesis.


## 3. Average Number of Goals

**Null Hypothesis (H0):**
The average number of goals scored by Barcelona is the same at Montjuic and Camp Nou.

**Alternative Hypothesis (H1):**
The average number of goals scored by Barcelona is different at Montjuic and Camp Nou.


```python
camp_nou_goals = df[df['stadium'] == 'Camp Nou']['goals_for']
montjuic_goals = df[df['stadium'] == 'Montjuic']['goals_for']

# Performing one-tailed t-Test
t_stat, p_val = stats.ttest_ind(montjuic_goals, camp_nou_goals)
print(f"Two-Tailed t-Test for Average Goals: t-stat = {t_stat}, p-value = {p_val}")
```

**Two-Tailed t-Test Result for Average Goals:**
- t-Statistic: 0.76
- p-Value: 0.45

**Interpretation:**
The p-value is greater than 0.05, suggesting that there is no statistically significant difference in the average number of goals scored between Montjuic and Camp Nou. Therefore, we fail to reject the null hypothesis.


## 4. Average xG

**Null Hypothesis (H0):**
The average xG for Barcelona is the same at Camp Nou and Montjuic.

**Alternative Hypothesis (H1):**
The average xG for Barcelona is different at Camp Nou and Montjuic.


```python
# Filtering data by stadium
camp_nou_xG_home = df[df['stadium'] == 'Camp Nou']['xG_home']
montjuic_xG_home = df[df['stadium'] == 'Montjuic']['xG_home']

# Performing two-tailed t-Test
t_stat_xG, p_val_xG = stats.ttest_ind(camp_nou_xG_home, montjuic_xG_home)
print(f"Two-Tailed t-Test for xG: t-stat = {t_stat_xG:.2f}, p-value = {p_val_xG:.2f}")
```

**Two-Tailed t-Test Result for Average xG:**
- t-Statistic: 0.57
- p-Value: 0.57

**Interpretation:**
The p-value is greater than 0.05, suggesting that there is no statistically significant difference in the average xG for Barcelona between Camp Nou and Montjuic. Therefore, we fail to reject the null hypothesis.


## 5. Average Goal Conceded

Null Hypothesis (H0):
The average number of goals scored against Barcelona is the same at Camp Nou and Montjuic.

Alternative Hypothesis (H1):
The average number of goals scored against Barcelona is different at Camp Nou and Montjuic.

```python
# Filtering data by stadium
camp_nou_goals_against = df[df['stadium'] == 'Camp Nou']['goals_against']
montjuic_goals_against = df[df['stadium'] == 'Montjuic']['goals_against']

# Performing two-tailed t-Test
t_stat_goals_against, p_val_goals_against = stats.ttest_ind(camp_nou_goals_against, montjuic_goals_against)
print(f"Two-Tailed t-Test for Goals Against: t-stat = {t_stat_goals_against:.2f}, p-value = {p_val_goals_against:.2f}")
```

**Two-Tailed t-Test Result for Goals Against Barcelona:**
- t-Statistic: -2.37
- p-Value: 0.02

**Interpretation:**
The p-value is less than 0.05, suggesting that there is a statistically significant difference in the average number of goals scored against Barcelona between Camp Nou and Montjuic. The negative t-statistic indicates that the average number of goals scored against Barcelona is higher at Montjuic than at Camp Nou. Therefore, we reject the null hypothesis.
